```python
import pandas as pd
```

```python
data = pd.read_csv('Data_training_and_val_set_CNvsAD.csv')
```

```python
data.head(10)
```

```python
'''
    Round data to 2 decimal places
'''
data.round({'LINFLATVEN': 2, 'RINFLATVEN': 2, 'LMIDTEMP': 2, 'RMIDTEMP': 2, 
           'LINFTEMP': 2, 'RINFTEMP': 2, 'LFUSIFORM': 2, 'RFUSIFORM': 2, 
           'LENTORHIN': 2, 'RENTORHIN': 2})
```

```python
'''
    Check for missing values
'''
data.isnull().sum()
```

```python
'''
    How many smaples are positive or negative?
'''
print("The count of the number of +ve and -ve samples are:")
print(data['DXCURREN'].value_counts())
```

```python
'''
    Mean values for positive samples
'''
pos_stats = data[data['DXCURREN']==1].describe().T
pos_stats = pos_stats[:-1]
pos_means = pos_stats['mean']
pos_means
```

```python
'''
    Mean values for negative samples
'''
neg_stats = data[data['DXCURREN']==0].describe().T
neg_stats = neg_stats[:-1]
neg_means = neg_stats['mean']
neg_means
```

```python
'''
    Compare percentage change between positive and negative samples
'''
diff_means = pd.DataFrame() # create the empty data frame

diff_means['features'] = neg_stats.index # add columns
diff_means['neg_means'] = neg_means.values
diff_means['pos_means'] = pos_means.values
perc = ((diff_means['pos_means'] - diff_means['neg_means']) / 
       diff_means['neg_means'])
diff_means['percentage_change'] = perc

diff_means
```

```python
'''
    P values to test whether positive and negative samples differ
'''
from scipy import stats

dataset = data # copy data

a = dataset[dataset['DXCURREN']==0].drop(columns=['DXCURREN'])
b = dataset[dataset['DXCURREN']==1].drop(columns=['DXCURREN'])

pvals = (pd.DataFrame(stats.ttest_ind(a,b)).T).drop(columns=[0]).rename(columns={1: "p-value"})
pvals
```

```python
'''
    Rank these features in decreasing p value
'''
from matplotlib import pyplot as plt

rank_pval = pd.DataFrame() # create data frame

rank_pval['feature'] = neg_stats.index

rank_pval['p_value'] = pvals['p-value']

# rank data frame by value
rank_pval = rank_pval.sort_values(by=['p_value'],ascending=False)

# plot ranking
rank_pval.plot.bar(x='feature', y='p_value', rot=0,figsize=(22,8), color='red')
plt.xticks(rotation='vertical',fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Feature', fontsize=14)
plt.ylabel('p_value',fontsize=14)
plt.title('Features Ranked by P-value',fontsize=20)
plt.axhline(y = 0.05, color = 'r', linestyle = '--')
plt.tight_layout()

rank_pval
```

```python
'''
    Indentify insignificant features
'''
insig_feat = rank_pval[rank_pval.p_value > 0.05]
print("The insignificant features of the dataset are:")
insig_feat.feature # only 1 insignificant feature
```

```python
'''
    Normalsise Data 
'''
df_min_max_scaled = data.copy()
  
# apply normalization techniques
for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  
# view normalized data
df_min_max_scaled
```

```python
'''
    Splitting data into training and test sets
'''
from sklearn.model_selection import train_test_split

df = df_min_max_scaled.copy()

# remove class column and insignificant features from training set
x = df.drop(columns = ['DXCURREN', 'EICV', 'RID'])
y = df['DXCURREN']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, 
                                                   random_state = 42)
```

```python
'''
    Logistic Regression Classifier 
'''
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression

lr_class = LogisticRegression(random_state = 42)

lr_class.fit(x_train, y_train)

lr_pred = lr_class.predict(x_test)

lr_cfm = confusion_matrix(y_test, lr_pred)

lr_TP=lr_cfm[1,1]
lr_TN=lr_cfm[0,0]
lr_FP=lr_cfm[1,0]
lr_FN=lr_cfm[0,1]

lr_Accuracy = (lr_TP+lr_TN)/(lr_TP+lr_TN+lr_FN+lr_FP)
lr_Sensitivity = lr_TP / (lr_TP + lr_FN)
lr_Specificity = lr_TN / (lr_TN + lr_FP)
lr_Precision = lr_TP / (lr_TP + lr_FP)
lr_Recall = lr_TP / (lr_TP + lr_FN)

print("Accuracy =",lr_Accuracy,", Sensitivity =",lr_Sensitivity,
      ", Specificity =",lr_Specificity, ", Precision =",lr_Precision, 
      ", Recall =",lr_Recall)

ConfusionMatrixDisplay(lr_cfm).plot() # plot confusion matrix
```

```python
'''
    Gaussian Naive-Bayes Classifier
'''
from sklearn.naive_bayes import GaussianNB

gnb_class = GaussianNB()

gnb_class.fit(x_train, y_train)

gnb_pred = gnb_class.predict(x_test)

gnb_cfm = confusion_matrix(y_test, gnb_pred)

gnb_TP=gnb_cfm[1,1]
gnb_TN=gnb_cfm[0,0]
gnb_FP=gnb_cfm[1,0]
gnb_FN=gnb_cfm[0,1]

gnb_Accuracy = (gnb_TP+gnb_TN)/(gnb_TP+gnb_TN+gnb_FN+gnb_FP)
gnb_Sensitivity = gnb_TP / (gnb_TP + gnb_FN)
gnb_Specificity = gnb_TN / (gnb_TN + gnb_FP)
gnb_Precision = gnb_TP / (gnb_TP + gnb_FP)
gnb_Recall = gnb_TP / (gnb_TP + gnb_FN)

print("Accuracy =",gnb_Accuracy,", Sensitivity =",gnb_Sensitivity,
      ", Specificity =",gnb_Specificity, ", Precision =",gnb_Precision, 
      ", Recall =",gnb_Recall)

ConfusionMatrixDisplay(gnb_cfm).plot() # plot confusion matrix
```

```python
'''
    Compare the model performances
'''
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# Classifier
classif = ['LogisticRegression', 'GaussianNB']

# Accuracy
lr_acc = accuracy_score(y_test, lr_pred)*100
gnb_acc = accuracy_score(y_test, gnb_pred)*100
accuracy = [lr_acc, gnb_acc]

# F1 score
lr_f1 = f1_score(y_test, lr_pred)
gnb_f1 = f1_score(y_test, gnb_pred)
score_f1 = [lr_f1, gnb_f1]

# Precision
lr_prec = precision_score(y_test, lr_pred)
gnb_prec = precision_score(y_test, gnb_pred)
precision = [lr_prec, gnb_prec]

# Recall
lr_rec = recall_score(y_test, lr_pred)
gnb_rec = recall_score(y_test, gnb_pred)
recall = [lr_rec, gnb_rec]

# Sensitivity
sensitivity = [lr_Sensitivity, gnb_Sensitivity]

# Specificity
specificity = [lr_Specificity, gnb_Specificity]

class_performance = pd.DataFrame()
class_performance['classifiers'] = classif
class_performance['accuracy'] = accuracy
class_performance['f1'] = score_f1
class_performance['precision'] = precision
class_performance['recall'] = recall
class_performance['sensitivity'] = sensitivity
class_performance['specificity'] = specificity

class_performance
```

```python
'''
    ROC curve Logistic Regression
'''
import matplotlib.pyplot as plt
from sklearn import metrics

y_pred_prob = lr_class.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_prob)
auc_lr = metrics.roc_auc_score(y_test, y_pred_prob)

plt.plot(fpr,tpr, label = "AUC = " + str(auc_lr))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.title('ROC Curve for Logistic Regression Model', fontsize = 12)
plt.show()
```

```python
'''
    ROC curve Gaussian Naive-Bayes
'''
y_pred_prob = gnb_class.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_prob)
auc_gnb = metrics.roc_auc_score(y_test, y_pred_prob)

plt.plot(fpr,tpr, label = "AUC = " + str(auc_gnb))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.title('ROC Curve for Gaussian Naive-Bayes Model', fontsize = 12)
plt.show()
```

```python
'''
    Comparing the AUC of each model
'''
lr_auc = auc_lr*100
gnb_auc = auc_gnb*100
print(lr_auc, gnb_auc)
d = gnb_auc - lr_auc
diff = d.round(2)
print("The GNB model is greater by",diff,"%")
```
